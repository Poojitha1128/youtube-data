# 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.


```python
#group by months and aggregate months on to grouped months
month_published=data.groupby("published_month")["video_id"].count().sort_values(ascending=False)
month_published
```

# 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
data[["view_count","dislikes"]].corr()
```

```python
plot.scatter(data["view_count"],data["dislikes"])
plot.xlabel("view_count")
plot.ylabel("dislikes")
plot.title("relation between view_count and dislikes")
plot.show()
```

```python
sns.regplot(data,x="view_count",y="dislikes")
```

### inference
here the correlation coefficient of view_count and dislikes is 0.68488. so there is positive correlation between both variables
so that if increase in view_count leads to increase in dislikes
